```python
import os
import numpy as np
import pandas as pd
import matplotlib.pylab as plt

import torch
from UAVIR_tools import BIRDSAIDataset, imgBoxes, frameVisualicer
from torchvision import transforms as T

###################################################
# IMPORTANTE: Instalar la base de datos siguiendo #
# las instrucciones del README e indicar en estas #
# dos variables los direcctorios!!                #
TRAIN_REAL_DATA_PATH = "D:\Datos\TrainReal"
TRAIN_SIM_DATA_PATH  = "D:\Datos\TrainSimulation"
TEST_DATA_PATH       = "D:\Datos\TestReal"
###################################################
MODEL_PATH = "model"

# Cargamos los dataset de interes
#dataset_trainSim = BIRDSAIDataset(TRAIN_SIM_DATA_PATH)
dataset_test = BIRDSAIDataset(TEST_DATA_PATH)
```

```python
N = 20

FIGSIZE = [16, 9]
RES = 2560 # 2k

# Mostramos un resumen del dataset
indices = torch.randperm(len(dataset_test)).tolist() 

fig = plt.figure(figsize=FIGSIZE, dpi=RES/FIGSIZE[0])
ax_list = fig.subplots(int(N/5), 5).flatten()
plt.subplots_adjust(wspace=0.1, hspace=0.1)

for i in range(N):
    img, target = dataset_test[indices[i]]
    img_box = imgBoxes(img, target["boxes"], target["labels"])
    
    ax_list[i].axis("off")
    ax_list[i].imshow(img_box)
```
